```python
def generate_tags_count_dict(data):
    tags = data['tags']
    tags_count_dict = {}
    for line in tags:
        line = line.lower()
        line = line.split("|")
        for word in line:
            word = re.sub("(\")","",word)
            if word not in tags_count_dict.keys():
                tags_count_dict[word] = 0
            tags_count_dict[word] = tags_count_dict[word] + 1
    return tags_count_dict
```

```python
def generate_channel_info_dict(data):
    channel_info = {}
    channel_info['count'] = len(data)
    channel_info['most_pop_video'] = data.sort_values(by='views', ascending=False).iloc[0]['video_id']
    channel_info['max_views'] = data.sort_values(by='views', ascending=False).iloc[0]['views']
    channel_info['total_views'] = sum(data['views'])
    channel_info['avg_views'] = data['views'].mean()
    channel_info['total_likes'] = sum(data['likes'])
    channel_info['avg_likes'] = data['likes'].mean()
    channel_info['total_dislikes'] = sum(data['dislikes'])
    channel_info['avg_dislikes'] = data['dislikes'].mean()
    channel_info['total_comment'] = sum(data['comment_count'])
    channel_info['avg_comment'] = data['comment_count'].mean()
    return channel_info
```

```python
def generate_channel_dict(data):
    channel_tags_dict = {}
    channel_video_dict = {}
    channel_category_dict = {}
    channel_info_dict = {}
    channel_list = set(data['channel_title'])
    for channel in channel_list:
        channel_data = data.groupby("channel_title").get_group(channel)
        channel_info_dict[channel] = generate_channel_info_dict(channel_data)
        channel_video_dict[channel] = channel_data['video_id'].tolist()
        channel_tags_dict[channel] = generate_tags_count_dict(channel_data)
        channel_category_dict[channel] = set(channel_data['category_id'])
    return channel_info_dict, channel_tags_dict, channel_video_dict, channel_category_dict
```

```python
channel_info_dict, channel_tags_dict, channel_video_dict, channel_category_dict = generate_channel_dict(us_data)
```
